# Visualize CAD UV Mapping Test Cases

This notebook displays the deterministic low/high CAD pairs used by the mapper tests. It keeps outputs empty in git; run the cells locally to open the `jupyter_cadquery` widget view.

````python
from pathlib import Path
import sys

repo_root = Path.cwd()
while repo_root != repo_root.parent and not (repo_root / "pyproject.toml").exists():
    repo_root = repo_root.parent

if not (repo_root / "pyproject.toml").exists():
    raise RuntimeError("Run this notebook from inside the cad-uv-map repository")

if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

print(repo_root)
````

````python
from jupyter_cadquery import show
import ipywidgets as widgets
from IPython.display import display

from tests.fixtures.cad_cases import (
    flat_to_u_groove_pair,
    flat_to_v_groove_pair,
    identity_box_pair,
    pedestal_ribs_nofold_pair,
    pedestal_ribs_pair,
)

CASES = {
    "identity_box": identity_box_pair,
    "flat_to_u_groove": flat_to_u_groove_pair,
    "flat_to_v_groove": flat_to_v_groove_pair,
    "pedestal_ribs": pedestal_ribs_pair,
    "pedestal_ribs_nofold": pedestal_ribs_nofold_pair,
}

def make_display_shapes(case_name):
    pair = CASES[case_name]()
    low = pair.low.translate((-18.0, 0.0, 0.0))
    high = pair.high.translate((18.0, 0.0, 0.0))
    return pair, low, high
````

Use the dropdown to inspect each low/high pair. The low object is shifted left; the high-detail object is shifted right.

````python
case_dropdown = widgets.Dropdown(
    options=list(CASES.keys()),
    value="flat_to_u_groove",
    description="Case",
)

def view_case(case_name):
    pair, low, high = make_display_shapes(case_name)
    print(f"{pair.name}: low faces={len(pair.low.faces())}, high faces={len(pair.high.faces())}")
    show(
        low,
        high,
        names=["low", "high"],
        colors=["#7aa6d8", "#d88f63"],
        alphas=[0.88, 0.88],
        axes=True,
        grid=True,
        render_edges=True,
    )

widgets.interact(view_case, case_name=case_dropdown);
````
